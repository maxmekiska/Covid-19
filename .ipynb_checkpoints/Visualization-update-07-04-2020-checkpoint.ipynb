```python
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv'
rdf = pd.read_csv(url) # recovered
```

```python
# function to extract data from git hub repo
def extractor(df,y):
    output = df.loc[df['Country/Region'] == y]
    return pd.DataFrame(output)
```

```python
# function to prepare data for further processing
def plotprep(df):
    numbers = df.iloc[:,4:-1]
    numbers2 = numbers.transpose()
    return numbers2
```

```python
def plotprep2(df):
    numbers = df.iloc[:,4:-1]
    numbers.columns = numbers.columns.astype(str) + '20'
    numbers.columns = pd.to_datetime(numbers.columns, format='%m/%d/%Y')
    numbers2 = numbers.transpose()
    return numbers2
```

```python
# importing and preparing world data
cdf1 = cdf.drop(['Province/State','Country/Region','Lat', 'Long'], axis = 1)
ddf1 = ddf.drop(['Province/State','Country/Region','Lat', 'Long'], axis = 1)
rdf1 = rdf.drop(['Province/State','Country/Region','Lat', 'Long'], axis = 1)

cdf1 = cdf.agg(['sum'])
ddf1 = ddf.agg(['sum'])
rdf1 = rdf.agg(['sum'])

cdf1 = cdf1.rename(index={'sum': 'cases'})
ddf1 = ddf1.rename(index={'sum': 'deaths'})
rdf1 = rdf1.rename(index={'sum': 'recoveries'})

world = pd.concat([cdf1, ddf1,rdf1])
world = plotprep(world)
```

```python
def gen_hist_ratio(data, name):
    place = []
    for i in range(data.shape[1]):
        place.append((data.iloc[:,i][1])/(data.iloc[:,i][0]))
    place = pd.DataFrame(place)
    place = place.rename(columns={0:name})
    return place
```

## Real estimate of death ratio (CFR)

#### Data is a pure estimation and is likely NOT representing the real situation

```python
def cfr(lag, data): # this is mathematical correct, please check again
    cases = data.iloc[:,0].diff().shift(lag)
    rate = (data.iloc[:,1].diff())/(cases)
    return rate
```

```python
def cfr2(data): 
    cases, deaths, recovery_rate = data.iloc[:,0], data.iloc[:,1], data.iloc[:,2] 
    ratio = (deaths)/(cases+recovery_rate)
    return ratio
```

## Real number of  infection estimation

#### Data is a pure estimation and is likely NOT representing the real situation

```python
def estcases(df, cfr, death_timeframe, days_to_double, decay_factor):
    days_to_double = days_to_double *(decay_factor+1)
    last_death = df.iloc[-1,1] # last death 
    calc =  ((last_death)/(cfr))*100 # calculating cases - (death_timeframe) days ago
    growth = ((death_timeframe//days_to_double*2))*(((death_timeframe%days_to_double)/days_to_double)+1) # calculating cases now
    estimate = growth * calc 
    return estimate
```

```python
world_est = estcases(world, 4, 7, 2,0) # assumption of doubling every two days
world_est
```

```python
prep_italy = plotprep2(composite_Italy)
italy_est = estcases(prep_italy, 4, 7, 3, 0)
italy_est
```

```python
def worstday(returns):
    out = returns.loc[returns == returns.min()]
    print('Date:', out.index[0],'Value:', returns.min()*100, '%')
```
